```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# 1. Load data set using pandas function and print last 5 records.

data = pd.read_csv("diamonds.csv")
data
```

```python
data.tail(5)
```

```python
# 2 Draw the bar chart of costliest and cheapest diamonds

dprice = data['price']
pmax = max(dprice)
pmin = min(dprice)
print(pmin,pmax)

```

```python
sorted_data = data.sort_values('price', ascending=False)

top_diamonds = sorted_data.head(1)
bottom_diamonds = sorted_data.tail(1)

selected_data = pd.concat([top_diamonds, bottom_diamonds])

selected_data.plot(x='carat', y='price', kind='bar', figsize=(8, 6), grid=True)

plt.title('Costliest and Cheapest Diamonds')
plt.xlabel('Carat')
plt.ylabel('Price')

plt.show()
```

```python
# 3 Draw the chart for Price vs Carat of Diamonds.

x = data["price"]
y = data["carat"]

plt.plot(x,y)
```
