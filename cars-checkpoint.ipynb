```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# 1. load data set using pandas function and print first 5 records.

data = pd.read_csv("cars.csv")
data
```

```python
# 3. find the mean price for each brand using pandas function.

mean = data.groupby("brand")["price"].mean()
mean


```

```python
# 4. Count the car having Black color

cars = (data["color"] == "black").count()
cars
```

```python
# 5. Find which Brand have maximum car in year 2017

d = data[data["year"] == 2017]
d

car = d.groupby('brand')['model'].count()

brand = car.idxmax()
brand
```

```python
# 6. Create a line chart which shows proportion for each car brand.

cars = data['brand'].unique()
num_car = []
for brand in cars:
    num = len(data[data['brand']==brand])
    num_car.append(num)
print(num_car)
fig = plt.figure(figsize =(10, 7))
plt.pie(num_car, labels = cars)
 
# show plot
plt.show()
```
